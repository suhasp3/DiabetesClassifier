```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
diabetes_large = pd.read_csv("diabetes.csv")
diabetes_large
```

```python
for feature in ['BMI', 'Age', 'GenHlth']:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=diabetes_large, x='Diabetes_012', y=feature)
    plt.title(f'Box Plot: {feature} by Diabetes Class')
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=diabetes_large, x='Age', y='BMI', hue='Diabetes_012')
plt.title('Line Plot: BMI vs Age by Diabetes Class')
plt.show()
```

```python
#Shows difference between diabetes and no_diabetes data 
no_diabetes_data = diabetes_large[diabetes_large['Diabetes_012'] == 0]
diabetes_data = diabetes_large[diabetes_large['Diabetes_012'] == 2]
print(len(no_diabetes_data))
print(len(diabetes_data))
```

```python

plt.figure(figsize=(8, 6))
sns.histplot(
    data=no_diabetes_data,
    x='GenHlth',
    bins=5,
    stat='probability',  # Normalize to proportions
    kde=False,
    color='blue'
)
plt.title('Proportion of Health Scores: No Diabetes')
plt.xlabel('General Health Rating (1=Excellent, 5=Poor)')
plt.ylabel('Proportion')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(
    data=diabetes_data,
    x='GenHlth',
    bins=5,
    stat='probability',  # Normalize to proportions
    kde=False,
    color='red'
)
plt.title('Proportion of Health Scores: Diabetes')
plt.xlabel('General Health Rating (1=Excellent, 5=Poor)')
plt.ylabel('Proportion')
plt.show()
```

```python

# Filter the data to include only No Diabetes (0) and Diabetes (2)
filtered_data = diabetes_large[diabetes_large['Diabetes_012'].isin([0, 2])]

# Create a grouped histogram
sns.histplot(
    data=filtered_data,
    x='GenHlth',
    hue='Diabetes_012',
    bins=5,
    kde=False,
    palette=['blue', 'green'],
    multiple='dodge'  # Ensures the bars are side-by-side
)

# Add labels and title
plt.title('Health Ratings Distribution: No Diabetes vs. Diabetes')
plt.xlabel('General Health Rating (1=Excellent, 5=Poor)')
plt.ylabel('Frequency')
plt.legend(title='Diabetes Class', labels=['No Diabetes', 'Diabetes'])
plt.show()

```

```python
key_features = ['GenHlth', 'HighBP', 'BMI', 'DiffWalk', 'HighChol', 'Age', 'Diabetes_012']
diabetes = diabetes_large[key_features]
diabetes
```

```python
def split_data(fraction): 
    global train_diabetes, test_diabetes
    train_diabetes = diabetes.sample(frac = fraction)
    test_diabetes = diabetes.drop(index=train_diabetes.index)
    print("train_diabetes:", train_diabetes.shape, "test_diabetes:", test_diabetes.shape)
split_data(0.8)
```

```python

X_train_diabetes = train_diabetes.drop(columns = ["Diabetes_012"])
y_train_diabetes = train_diabetes['Diabetes_012']
X_test_diabetes = test_diabetes.drop(columns = ["Diabetes_012"])
y_test_diabetes = test_diabetes['Diabetes_012']
```

```python
T = DecisionTreeClassifier(max_depth=3, class_weight='balanced')
T.fit(X_train_diabetes, y_train_diabetes)
train_score = T.score(X_train_diabetes, y_train_diabetes)
test_score = T.score(X_test_diabetes, y_test_diabetes)
print('Score on train:', train_score)
print('Score on test:', test_score)

fig, ax = plt.subplots(1, figsize = (20, 20))
p = plot_tree(T, filled = True, feature_names = X_train_diabetes.columns)
```
