```python
key_features = ['GenHlth', 'HighBP', 'BMI', 'DiffWalk', 'HighChol', 'Age', 'Diabetes_012']
diabetes = diabetes_large[key_features]
diabetes
```

```python
def split_data(fraction): 
    global train_diabetes, test_diabetes
    train_diabetes = diabetes.sample(frac = fraction)
    test_diabetes = diabetes.drop(index=train_diabetes.index)
    print("train_diabetes:", train_diabetes.shape, "test_diabetes:", test_diabetes.shape)
split_data(0.8)
```

```python

X_train_diabetes = train_diabetes.drop(columns = ["Diabetes_012"])
y_train_diabetes = train_diabetes['Diabetes_012']
X_test_diabetes = test_diabetes.drop(columns = ["Diabetes_012"])
y_test_diabetes = test_diabetes['Diabetes_012']
```

```python
T = DecisionTreeClassifier(max_depth=3, class_weight='balanced')
T.fit(X_train_diabetes, y_train_diabetes)
train_score = T.score(X_train_diabetes, y_train_diabetes)
test_score = T.score(X_test_diabetes, y_test_diabetes)
print('Score on train:', train_score)
print('Score on test:', test_score)

fig, ax = plt.subplots(1, figsize = (20, 20))
p = plot_tree(T, filled = True, feature_names = X_train_diabetes.columns)
```
