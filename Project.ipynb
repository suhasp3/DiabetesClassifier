```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.utils import resample
from sklearn.model_selection import train_test_split

```

```python
#
def feature_boxplot(feature, BMI = False):
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=diabetes_large, x='Diabetes_012', y=feature)
    plt.title(f'Box Plot: {feature} by Diabetes Class')
    if BMI == True:
        plt.axhline(25, color="red", linestyle="--")
        plt.text(x=-0.75,y=22,s='Healthy BMI',color="red",fontsize=10,ha='center',va='bottom')
    plt.show()
```

```python
key_features = ['GenHlth', 'HighBP', 'BMI', 'DiffWalk', 'HighChol', 'Age', 'Diabetes_012']
diabetes = diabetes_large[key_features]
diabetes
```

```python
no_diabetes = diabetes[diabetes['Diabetes_012'] == 0]
pre_diabetes = diabetes[diabetes['Diabetes_012'] == 1]
yes_diabetes = diabetes[diabetes['Diabetes_012'] == 2]
print("Amount of entries with no diabetes in dataset: " + str(len(no_diabetes)))
print("Amount of entries with prediabetes in dataset: " + str(len(pre_diabetes)))
print("Amount of entries with diabetes in dataset: " + str(len(yes_diabetes)))
```

```python
resampled_pre_diabetes = resample(pre_diabetes, 
                                 replace=True,    
                                 n_samples= 213703)
                                 
resampled_yes_diabetes = resample(yes_diabetes, 
                                 replace=True,    # sample with replacement
                                 n_samples= 213703)
print("Amount of entries with no diabetes in train: " + str(len(no_diabetes)))
print("Amount of entries with prediabetes in train: " + str(len(resampled_pre_diabetes)))
print("Amount of entries with diabetes in train: " + str(len(resampled_yes_diabetes)))
```

```python
diabetes = pd.concat([no_diabetes, resampled_pre_diabetes, resampled_yes_diabetes])
diabetes.shape
```

```python
diabetes
```

```python
def split_data(fraction): 
    global train_diabetes, test_diabetes
    train_diabetes = diabetes.sample(frac = fraction)
    test_diabetes = diabetes.drop(index=train_diabetes.index)
    print("train_diabetes:", train_diabetes.shape, "test_diabetes:", test_diabetes.shape)
split_data(0.8)
```

```python
train_diabetes.describe()
```

```python
test_diabetes.describe()
```

```python

X_train_diabetes = train_diabetes.drop(columns = ["Diabetes_012"])
y_train_diabetes = train_diabetes['Diabetes_012']
X_test_diabetes = test_diabetes.drop(columns = ["Diabetes_012"])
y_test_diabetes = test_diabetes['Diabetes_012']
```

```python
T = DecisionTreeClassifier(max_depth=4)
T.fit(X_train_diabetes, y_train_diabetes)
train_score = T.score(X_train_diabetes, y_train_diabetes)
test_score = T.score(X_test_diabetes, y_test_diabetes)
print('Score on train:', train_score)
print('Score on test:', test_score)

fig, ax = plt.subplots(1, figsize = (20, 20))
p = plot_tree(T, filled = True, feature_names = X_train_diabetes.columns)
```

```python
train_scores=[]
test_scores=[]
r=range(1,40)
for i in r:
    T = DecisionTreeClassifier(max_depth=i) 
    T.fit(X_train_diabetes, y_train_diabetes)
    train_scores.append(T.score(X_train_diabetes, y_train_diabetes))
    test_scores.append(T.score(X_test_diabetes, y_test_diabetes))
fig, ax = plt.subplots(1)
sns.lineplot(x=r,y=train_scores,label='train')
sns.lineplot(x=r,y=test_scores,label='test')
ax.set_ylabel('Prediction score', fontsize=14)
ax.set_xlabel('Max tree depth', fontsize=14)
ax.legend()
```
